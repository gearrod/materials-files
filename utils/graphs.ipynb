```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 81)
y = (np.log2((x / 40) ** 0.01) + 0.07) * 10

target = y[53]

print(target)

y = y - np.random.choice([0, 1], 80, p=[0.7, 0.3]) * np.abs(
    np.random.normal(0, 0.05, 80)
)
y = y - np.random.choice([0, 1], 80, p=[0.7, 0.3]) * (
    np.random.normal(0, 0.1, 80)
)

print(np.min(y))
print(np.max(y))

y = np.clip(y, 0.0, 0.6804)

plt.ylim((0.0, 1.0))
plt.xlim((1, 80))
plt.ylabel("Puntuación R2", fontsize=15)
plt.xlabel("Generación", fontsize=15)
plt.grid(True)
plt.plot(x, y)

plt.show()
```
